## Finding Elements in Web Page using Python

### Finding Elements by Class

```python
class_data = soup.find("div", class_ = "container flex items-center h-full")      # "font-subtitle2 text-dark-label text-ellipsis-multiline-200 text-left"
```

```python
print(class_data)
```

```python
class_data.find_all('p')
```

### Finding Elements by id

```python
id_data = soup.find('nav', id = "nav")
```

```python
print(id_data)
```

```python
id_data.find_all('p')
```

## Extracting Text from the Tags in Web Page using Python

```python
web1 = requests.get("https://www.tutorialsfreak.com/")
```

```python
soup1 = BeautifulSoup(web1.content, "html.parser")
```

```python
print(soup1.prettify())
```

```python
lines = soup1.find_all("p")
```

```python
for l in lines:
    print(l.text)
```

```python
s = soup1.find('div', class_='d-md-flex align-items-md-center why-choose-left-section')
```

```python
lines_1 = s.find_all('p')
```

```python
for l1 in lines_1:
    print(l1.text)
```

## Extracting Links in Web Page using Python

```python
soup1.find_all('a')
```

```python
for a in soup1.find_all("a"):
    print(a.get("href"))
```

## Extracting Image Information in Web Page using Python

```python
soup1.find_all("img")
```

```python
for i in soup1.find_all("img"):
    print(i.get("src"))
```

```python
for i in soup1.find_all("img"):
    print(i.get("alt"))
```
